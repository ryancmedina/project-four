```python
import requests
from datetime import datetime
import json

# Set your parameters
api_key = 'aeRc3h5tR5wweegUs1yyQ98QGs1H4ti2jpGW5hGn'  # Replace with your own key if you have one
start_date = '2015-01-01'
end_date = '2024-12-31'
url = 'https://api.nasa.gov/DONKI/FLR'

# Build the query parameters
params = {
    'startDate': start_date,
    'endDate': end_date,
    'api_key': api_key
}

# Make the request
response = requests.get(url, params=params)

# Check for a successful response
if response.status_code == 200:
    data = response.json()
    print("Received data:")
    for flare in data:
        print(f"ID: {flare.get('flrID')}, Start Time: {flare.get('beginTime')}")
else:
    print(f"Failed to retrieve data: {response.status_code}")

# Local File Path
local_file_path = "../Resources/solar_flares.json"

# Convert DataFrame to a list of dictionaries
data_to_save = data.to_dict(orient='records')

# Save to local JSON file
try:
    with open(local_file_path, 'w') as f:
        json.dump(data, f, indent=4)
    print(f"Data successfully saved to {local_file_path}")
except Exception as e:
    print(f"Error saving to local file: {e}")
```
